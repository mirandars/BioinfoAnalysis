```python

with open('./data/TRIM5ALPHA_complete_PopSet', 'r') as myfile:
    data=myfile.read()
```

```python
data = data.split('>')
dic = {}
```

```python
for i in range(len(data)):
    data[i] = data[i].split('\n',maxsplit=1)
    if len(data[i]) > 1:
        dic[data[i][0]] = data[i][1].replace('\n','').split()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(dic)
```

```python
df = df.transpose()
```

```python
df.columns = ['Sequence']
```

```python
df['Length'] = df['Sequence'].apply(lambda seq: len(seq))
```

```python
df = df.reset_index()
```

```python
df
```

```python
df['Species'] = df['index'].apply(lambda name: name.split(' ')[1] + ' ' + name.split(' ')[2])
df['Adenine'] = df['Sequence'].apply(lambda seq: seq.count('A')/len(seq))
df['Cytosine'] = df['Sequence'].apply(lambda seq: seq.count('C')/len(seq))
df['Guanine'] = df['Sequence'].apply(lambda seq: seq.count('G')/len(seq))
df['Thymine'] = df['Sequence'].apply(lambda seq: seq.count('T')/len(seq))
df['Uracil'] = df['Sequence'].apply(lambda seq: seq.count('U')/len(seq))
df['Purine'] = df['Sequence'].apply(lambda seq: (seq.count('A')+ seq.count('G'))/len(seq))
df['Pyrimidines'] = df['Sequence'].apply(lambda seq: (seq.count('C')+ seq.count('T')+seq.count('U'))/len(seq))
df['Non-identified'] = df['Sequence'].apply(lambda seq: seq.count('N')/len(seq))

```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
df = df.set_index('Species')
```

```python
ax = df[['Adenine','Cytosine','Guanine','Thymine']].plot(kind='bar', title ="V comp", figsize=(15, 10), legend=True, stacked=True,fontsize=12)
```
